```python
#输入十个数 排序
num_list=[]
for i in range(10):
    num=int(input("请输入一个整数："))
    num_list.append(num)
print("数字列表：",num_list)
num_list.sort();
print("排序后数字列表：",num_list)
print("最大值为：",num_list[9])
print("最小值为：",num_list[0])
sum=0
for i in range(10):
    sum+=num_list[i]
ave=sum/10
print("平均值为：",ave)

```

```python
#合并两个列表 结果去重
nums_list1=[1,3,5,7,9]
nums_list2=[2,4,6,8,7]
for num in nums_list2:
    nums_list1.append(num)
nums_list3=[]
for num in nums_list1:
    if num not in nums_list3:
        nums_list3.append(num)
nums_list3.sort()        
print(nums_list3)        

```

```python
#生成1～20 平方列表
num_list=[i**2 for i in range(1,20)]
print(num_list)
```

```python
#提取偶数 输出平方
num_list=[i for i in range(1,20)]
num_list1=[num**2 for num in num_list if num%2==0]
print(num_list1)

```
